#  Data preparation

## Removing Duplicates

```python
# Checking duplicates
duplicates = df[df.duplicated()]
print("Duplicate Rows : ",len(duplicates))
duplicates
```

```python
# Removing duplicate rows from the dataset
df.drop_duplicates(inplace = True)
```

```python
df.shape
```

## Handling inbalance data

```python
df['Diabetes_012'].value_counts()
```

There is a huge Imbalance in the dataset. We need to fix this.

### We will use oversampling for minority classes to fix this imbalance.

```python
non_dia= df[df['Diabetes_012'] == 0]
pre_dia = df[df['Diabetes_012'] == 1]
dia = df[df['Diabetes_012'] == 2]

pre_dia_os = pre_dia.sample(len(non_dia), replace=True)
dia_os = dia.sample(len(non_dia), replace=True)

df_new = pd.concat([pre_dia_os,dia_os, non_dia], axis=0)
```

```python
df_new['Diabetes_012'].value_counts()
```

```python
sns.countplot(df_new['Diabetes_012'])
```

### Slicing Dataset



```python

# input
x = df_new.iloc[:, 1:].values
 
# output
y = df_new.iloc[:, 0].values
```

### Train_test_split

```python
#Spliting the data in 80:20 training to testing
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.20, random_state = 0)
```

### Scaling Dataset

```python
#Feature Scaling
#Applying MinMaxScaler
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

# Applying Machine learning models from Scikit-learn

## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state = 0)
classifier.fit(x_train, y_train)
```

```python
y_pred = classifier.predict(x_test)
```

```python
#Performace Evaluation
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
 
print ("Confusion Matrix : \n", cm)
```

```python
from sklearn.metrics import plot_confusion_matrix
plot_confusion_matrix(classifier,x_test,y_test,cmap='Blues')
plt.grid(False)
```

```python
from sklearn.metrics import accuracy_score
print ("Accuracy : ", accuracy_score(y_test, y_pred))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

##  Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
gnb.fit(x_train, y_train)
```

```python
y_pred_nb = gnb.predict(x_test)
```

```python
# comparing actual response values (y_test) with predicted response values (y_pred)
print("Gaussian Naive Bayes model accuracy(in %):", accuracy_score(y_test, y_pred_nb)*100)
```

```python
#Performace Evaluation
cm = confusion_matrix(y_test, y_pred_nb)
 
print ("Confusion Matrix : \n", cm)
```

```python
plot_confusion_matrix(gnb,x_test,y_test,cmap='Blues')
plt.grid(False)
```

```python
print(classification_report(y_test,y_pred_nb))
```

## Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
 
model_1 = RandomForestClassifier(n_estimators = 300, criterion = 'entropy',
                             min_samples_split=10, random_state=0)

# fitting the model on the train data
model_1.fit(x_train, y_train)

# predicting values on test data
predictions = model_1.predict(x_test)
```

```python
print("Random forest model accuracy(in %):", accuracy_score(y_test, predictions)*100)
```

```python
cm = confusion_matrix(y_test, predictions)
 
print ("Confusion Matrix : \n", cm)
```

```python
plot_confusion_matrix(model_1,x_test,y_test,cmap='Blues')
plt.grid(False)
```

```python
print(classification_report(y_test,predictions))
```

## LDA(Linear Discriminant Analysis)

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA

lda = LDA(n_components=1)
lda.fit(x_train, y_train)
# predicting values on test data
predictions_lda = lda.predict(x_test)
```

```python
print("LDA model accuracy(in %):", accuracy_score(y_test, predictions_lda)*100)
```

```python
plot_confusion_matrix(lda,x_test,y_test,cmap='Blues')
plt.grid(False)
```

```python
print(classification_report(y_test,predictions_lda))
```

## QDA(Quadratic Discriminant Analysis)

```python
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis as QDA

qda = QDA()
qda.fit(x_train, y_train)
# predicting values on test data
predictions_qda = qda.predict(x_test)
```

```python
print("QDA model accuracy(in %):", accuracy_score(y_test, predictions_qda)*100)
```

```python
plot_confusion_matrix(qda,x_test,y_test,cmap='Blues')
plt.grid(False)
```

```python
print(classification_report(y_test,predictions_qda))
```
